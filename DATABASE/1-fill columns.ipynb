```sql
UPDATE uber_data_final
SET requested_timestamp = DATEADD(MINUTE, (ABS(CHECKSUM(NEWID()))) % 26 - 40, tpep_pickup_datetime);
```

```sql
select 
max(DATEDIFF(MINUTE, requested_timestamp, tpep_pickup_datetime)),
min(DATEDIFF(MINUTE, requested_timestamp, tpep_pickup_datetime))
FROM uber_data_final;
```

```sql
UPDATE uber_data_final
SET accepted_timestamp = DATEADD(MINUTE, (ABS(CHECKSUM(NEWID()))) % 10 + 1, requested_timestamp);
```

```sql
select 
max(DATEDIFF(MINUTE, accepted_timestamp, requested_timestamp)),
min(DATEDIFF(MINUTE, accepted_timestamp, requested_timestamp)),
max(DATEDIFF(MINUTE, accepted_timestamp, tpep_pickup_datetime)),
min(DATEDIFF(MINUTE, accepted_timestamp, tpep_pickup_datetime))
FROM uber_data_final;
```
