```sql
select passenger_name, COUNT(*) as passenger_trips
FROM uber_data_final u 
GROUP BY passenger_name
order by passenger_trips desc;
```

```sql
select passenger_name, passenger_email,passenger_phone, COUNT(*) as passenger_trips
FROM uber_data_final u 
GROUP BY passenger_name, passenger_email, passenger_phone
order by passenger_trips desc;
```

```sql
SELECT 
    udf.passenger_id,
    udf.passenger_name, 
    udf.passenger_email, 
    udf.passenger_phone,
    udf2.driver_id,
    udf2.driver_name, 
    udf2.driver_email, 
    udf2.driver_phone
FROM uber_data_final udf
JOIN uber_data_final udf2 
    ON udf.passenger_id = udf2.driver_id
--WHERE udf.passenger_email = udf2.driver_email
--   OR udf.passenger_phone = udf2.driver_phone;
```
